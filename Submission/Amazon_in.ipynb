```python
import requests, bs4, pandas as pd, numpy as np
from bs4 import BeautifulSoup
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.75 Safari/537.36'}
```

```python
asin = []
brand = []
mobile = []
ratings = []
customer = []
reviews = [[]]
urls = []
for i in range(2,11):
    urls.append('https://www.amazon.in/s?k=Mobiles&i=electronics&rh=n%3A1389432031%2Cp_89%3AApple%7CMI%7COnePlus%7COppo%7CRedmi%7CSamsung%7CVivo%7CXiaomi%7Crealme&dc&page='+str(i)+'&qid=1602750419&rnid=3837712031&ref=sr_pg_'+str(i))
urls
```

```python
href = []
for url in urls:
    page = requests.get(url, headers = header)
    soup = BeautifulSoup(page.content,'html.parser')
    soup.prettify()
    names = soup.find_all('a',class_="a-link-normal a-text-normal")
    if(len(names)!=0):
        for i in range(4):
            names.pop(0)
    href.append(names)
    print(len(names))
href
```

```python
hrefs = []
for i in range(6):
    for j in range(22):
        hrefs.append(href[i][j].get('href'))
```

```python
customers = []
userid = []
reviews = []
date_of_review = []
ratings = []
model = []
asin = []
for ref in hrefs:
    url = 'https://www.amazon.in'+ref
    page = requests.get(url, headers = header)
    soup = BeautifulSoup(page.content,'html.parser')
    soup.prettify()
    uid = soup.find_all('div', {'data-hook':"genome-widget"})
    asin_no = soup.find(id='averageCustomerReviews')
    rating_no = soup.find_all('i', {'data-hook':'review-star-rating'})
    dor = soup.find_all('span', {'data-hook':'review-date'})
    Model_name = soup.find(id='productTitle').get_text().strip()
    cust_name = soup.find_all(class_='a-profile-name')
    review = soup.find_all('div', class_='a-expander-content reviewText review-text-content a-expander-partial-collapse-content')
    for i in range(len(rating_no)):
        userid.append(uid[i].find('a').get('href').split('/')[3].split('.')[2])
        model.append(Model_name.split(")")[0] + ')')
        date_of_review.append(dor[i].get_text().strip().split('on ')[1])
        asin.append(asin_no.get('data-asin'))
        customers.append(cust_name[i].get_text().strip())
        reviews.append(review[i].find('span').get_text().strip())
        ratings.append(float(rating_no[i].find('span').get_text().strip()[0]))
asin, model, customers, reviews, ratings, date_of_review, userid
```

```python
dict_amaz = {
    'user_id':userid,
    'ASIN':asin,
    'model_name':model,
    'customer_name':customers,
    'Review_date':date_of_review,
    'Reviews':reviews,
    'Ratings':ratings
}
```

```python
df = pd.DataFrame(dict_amaz)
df.to_csv('Amazon_in.csv')
```
